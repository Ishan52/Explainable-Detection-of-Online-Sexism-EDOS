```python
import pandas as pd
```

```python
original_df = pd.read_csv('Original Data.csv')
original_df
```

```python
training_df = original_df.sample(frac=0.8, random_state=1)
training_df
```

```python
testing_df = original_df.drop(training_df.index)
testing_df
```

```python
training_df.to_csv('Training Data.csv', index=False)
testing_df.to_csv('Testing Data.csv', index=False)
```
